```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['Severity', 'Start_Time', 'Weather_Condition', 'State']

df = pd.read_csv(
    "US_Accidents_March23.csv",
    usecols=cols,
    nrows=500000
)

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
plt.figure()
df['Severity'].value_counts().plot(kind='bar')
plt.title("Accident Severity Distribution")
plt.show()
```

```python
plt.figure()
df['State'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 States with Highest Accidents")
plt.show()
```

```python
plt.figure()
df['Weather_Condition'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Weather Conditions Causing Accidents")
plt.show()
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

plt.figure()
df['Start_Time'].dt.hour.value_counts().sort_index().plot(kind='line')
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```
